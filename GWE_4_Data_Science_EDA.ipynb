## Import Library

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = pd.read_csv('tips.csv')
```

```python
tips.tail(7)
```

## Business Understanding

Contoh Case Business:

Seorang pramusaji mencatat informasi tentang setiap tip yang dia terima selama beberapa bulan bekerja di satu restoran. Secara keseluruhan dia mencatat 244 tip.

## Data Understanding

```python
tips.head()
```

* total_bill : tagihan dalam dolar
* tip : tip dalam dolar
* sex : jenis kelamin pembayar tagihan
* smoker : apakah ada perokok di pesta itu
* day : hari 
* time : waktu makan
* size : ukuran party

##.sort_values() -> continues / nuemrik & kategorik -> ordinal

```python
tips.sort_values(by='total_bill', ascending=False)
```

##.groupby() -> diskrit / kategorik

```python
tips.groupby('sex').agg({'tip' : 'mean'})
# note : operasi bisa pakai count, mean, median, dan sebagainya
```

# Distribution Plot

## Distplot / Histogram

```python
sns.distplot(tips['total_bill'])
```

```python
sns.distplot(tips['total_bill'], kde=False, bins=50)
```

```python
sns.displot(tips, x="total_bill", kde=True, hue="sex")
```

## Jointplot

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='scatter', hue='sex', palette='rainbow')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips, kind='hex')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips, kind='reg')
```

# Pairplot

```python
sns.pairplot(tips, palette='coolwarm')
```

```python
sns.pairplot(tips,hue='sex',palette='coolwarm')
```

# Categorical Plots

## barplot and countplot

```python
sns.barplot(x='day', y='total_bill',data=tips)
```

```python
sns.countplot(x='day',data=tips)
```

## boxplot and violinplot


```python
sns.boxplot(x="day", y="total_bill", data=tips, palette='rainbow')
```

```python
# Can do entire dataframe with orient='h'
sns.boxplot(data=tips,palette='rainbow', orient='h')
```

```python
sns.boxplot(x="day", y="total_bill", hue="smoker", data=tips, palette="coolwarm")
```

```python
sns.violinplot(x="day", y="total_bill", data=tips ,palette='rainbow')
```

```python
sns.violinplot(x="day", y="total_bill", data=tips, hue='sex',palette='Set1')
```

```python
sns.violinplot(x="day", y="total_bill", data=tips,hue='sex', split=True,palette='Set1')
```

# Matrix Plot

## Heatmap

```python
tips.corr()
```

```python
sns.heatmap(tips.corr())
```

```python
sns.heatmap(tips.corr(),cmap='rainbow', annot=True)
```

```python
flights = sns.load_dataset('flights')
flights.head()
```

```python
pvflights = flights.pivot_table(values='passengers',index='month',columns='year')

sns.heatmap(pvflights)
```

```python
sns.heatmap(pvflights,cmap='coolwarm',linecolor='white',linewidths=1)

plt.title('\nRiwayat Penerbangan\n', size=17, color='red')
plt.xlabel("\nTahun", size=12)
plt.ylabel("\nTahun", size=12)
```

# Regression Plot

```python
sns.regplot(x='total_bill',y='tip',data=tips)
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips, hue='sex')
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips, hue='smoker', palette='magma')
```
